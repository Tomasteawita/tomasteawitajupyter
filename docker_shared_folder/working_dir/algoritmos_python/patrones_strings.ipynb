````python
def merge_lists(nums1, m, nums2, n):
   # Tu c칩digo aqu칤 游녢
   z = m + n
   while m != 0 or n != 0:
      if m == 0:
         nums1[z-1] = nums2[n-1]
         n -= 1
      elif n == 0 or nums1[m-1] > nums2[n-1]:
         nums1[z-1] = nums1[m-1]
         m -= 1
      else:
         nums1[z-1] = nums2[n-1]
         n -= 1
      z -= 1
   return nums1
    


````

# Containerwith most water
Dada una lista de n칰meros que representa un gruoi de lineas de diferentes alturas.
Encuentra dos lineas que formen un contenddor , tal que este contenga lamayorcantidad de agua.
Devuelve la cantidad m치xima de agua que puede almacenar un contenedor.

Ejemplo visual de como se ver칤a el ejercicio a resolver:


Datos de pruebas:
[1,8,6,2,5,4,8,3,7]

````python
def maxArea(heights):
    lp = 0 # left pointer
    rp = len(heights)-1 # right pointer
    maxArea = -1

    while lp < rp:
        # compute area
        area = (rp -lp) * min(heights[lp], heights[rp])
        # get max. area
        maxArea = max(area, maxArea)
        # discard smallest value
        if heights[lp] < heights[rp]:
            lp += 1
        else:
            rp -= 1
    return maxArea

heights = [1,8,6,2,5,4,8,3,7]
print(maxArea(heights))

````
