Como Scala mezcla conceptos de programación orientada a objetos con conceptos de programación funcional, el polimorfismo generado por la herencia de estos tipos de datos puede ser importante en algunas ocasiones, aunque no te afectará para los temas que estaremos tratando a lo largo del curso.

Crear un **grupo de funciones** para hallar la media, la mediana y la moda de una lista de números devolviendo la respuesta dentro de un tipo Map.

```scala
val fivesScala = List(5.0, 15.0, 20.0, 25.0)
```

```scala
def media(nums: List[Double]): Double = nums.sum / nums.length

def mediana(nums: List[Double]): Double = {
  val sortedNums = nums.sorted
  val middle = nums.length / 2
  if (nums.length % 2 == 0) (sortedNums(middle - 1) + sortedNums(middle)) / 2.0
  else sortedNums(middle)
}

def moda(nums: List[Double]): Double = {
  nums.groupBy(identity).mapValues(_.size).maxBy(_._2)._1
}

def calcular(nums: List[Double]): Map[String, Double] = {
  Map("Media" -> media(nums), "Mediana" -> mediana(nums), "Moda" -> moda(nums))
}
```

## ¿Qué es un case class y en que se diferencia con la class en Scala?
un `case class` es una clase especial que se utiliza comunmente para modelar los datos inmutables. Aunque comparte muchas características con las  clases normales, **hay algunas diferencias clasve para que los case class sean utiles en escenarios especificos**
1. **Inmutabilidad por defecto**: Los `case class` son inmutables por defecto, lo que sifnidica que los objetos de este tipo no pueden ser modificados después de que se han creado. Esto facilita el razonamietno acerca de los datos y ayuda a prevenir efectos secundarios no deseados.
2. **Generación automática de métodos útiles**: Scala automáticamente genera metodo útiles para las `case class`, como metodos `toString`, `equals`, y `hashCode`. Esto es útil para imprimir representaciones legibles, comprar objetos y usar instancias de la clase en estructuras de datos que dependen de `hashCode`.
3. **Descomposición de patrones**: Los case class se utilizan frecuentemente con la descomposición de patrones en Scala. Esto permite escribir codigo conciso y expresivo para manipular y analizar la estructura de los objetos.
4. **Sintaxis concisa para la creación de instancas**: Puedes crear instancias ``vase ckass` sin utilizar la palabra clave `new`, lo que proporciona una sintaxis más concisa y legible.

A contuniación, se presenta un ejemplo suimple para ilustrar la diferencia entre una clase regultasra y una `case class`;

```scala
// Clase regular
class Persona(nombre: String, edad: Int)

// Case class
case class PersonaCase(nombre: String, edad: Int)
```

Con una `case class`, automáticamente obtienes métodos como "toString", 'equals', y 'hashCode' sin tener que escribirlos explícitamente También puede utilizar la descomposición de patrones de manera más efectiva:

```scala
val persona1 = PersonaCase("Juan", 30)

persona1 match {
  case PersonaCase("Juan", edad) => println(s"Encontré a Juan con edad $edad")
  case _ => println("No es Juan")
}
```

```scala
// Otros ejemplos
case class People(name: String, age: Int)

val p1 = People("Tomas", 20)
val p2 = People("Santiago", 15)

def h(x: People) = x match{
    case y if y.age >= 18 => "Mayor de edad"
    case _ => "No es mayor de edad"
}
```

# Recursión
Una manera de abordar los problemas donde la función se llama a sí misma.
Es la manera clásica de proramar en lenguajes funcionales.

## Tail recursión
Es una optimizaación de estos algoritmos

```scala
// EL factorial con recursión

def factorial(n: Long): Long = 
    if (n == 0) {
        println("---Finalizado---")
        1
    } else {
        n * factorial(n-1)
    }
```

```scala
val startTimeRecursion = System.nanoTime()
println(factorial(100))
val endTimeRecursion = System.nanoTime()
val duration = (endTimeRecursion - startTimeRecursion) / 1e6  // Convertir a milisegundos
println(s"Tiempo de ejecución: $duration ms")
```

## Inconvenientes solo con recursión
Hacerlo solo con recursión lo hace más ineficiente.
En la definición anterior, primero resta todos los numeros para al final hacer todas las multiplicaciónes.
Esto es más ineficientes ya que esta multiplicación no se va a realizar hasta que no se llegue al último numero, solo hastsa que llega hasta a 1, ahí arranca la multiplicación, por lo que los datos van a quedar almacecnados en memoria.

```scala
// Tail recursion

def factorialRecursion(n: Long, resultado: Long =  1L): Long = 
    if (n == 0) {
        println("---Finalizado---")
        resultado
    } else {
        factorialRecursion(n-1, n * resultado)
    }
```

```scala
val startTimeTailRecursion = System.nanoTime()
println(factorialRecursion(100))
val endTimeTailRecursion = System.nanoTime()
val duration = (endTimeTailRecursion - startTimeTailRecursion) / 1e6  // Convertir a milisegundos
println(s"Tiempo de ejecución: $duration ms")
```

# Razonamiento con tipos
El razonamiento inductivo es una manera de razonar donde el primer paso es encontrar lo que se conoce como caso base, es decir, uin hecho concreto que usaremos como inicio del proceso que queremos modelar.

El siguiente paso es modelar la continuacióon del proceso (paso inductivo) Usando elcaso base anterior. Lo que sucede entonces es que el resultado de este proceso se convierte en el nuevo caso base.
Esto puede repetirse cuantas veces sea necesario ya que conceptualmente puede ser infinito, aunque lo normal es hallar una condicion que detiene el proceso de continuación y entrega el resultado final.

## Ejemplo de razonamiento inductivo
Un ejemplo sencillo de una prueba por uunducción matemática sería probar la siguiente proposición:
"Los numero son infinitos".
* El caso base sería el número más bajo, que es el 0 (cero).
* La continuación se usa el caso base y se le suma el numero 1.
* La prueba es que, si para cualquier caso base puedo sumarle 1, entonces los números son infinitos.

El restulado de esta proposición es que es cuerta, porque a nivel matemático no hay nada que evite que puesa sumarle un 1 a cualquier otro número. Por supuesto esto no es cierto en informática, donde estamos limitados por el espacio en memoria, esto nos pone un límite donde ya no es posible sumar más cuando hay ún numero muy grande.

Al momento de pensar en la menera de solucionar un problema de programación, entrenar a nuestro cerebro paar usar el rezonamieto inductivo nos puede dar soluciones más elegantes y consistentes, lo que normalmente da como resultado software más "correcto".

Por su puesto, no en todos los casos una solución de esta naturaleza será la más eficiente computacionalmete, pero pasar de una solución inductiva a una que no, suele ser más facil que al revés.

## Proposiciones y pruebas
* los tupos, son proposiciones.
* Los programas, son pruebas.
Nuestros programas son como una gran función matemática.
El compulador se convierte en un razonador que prueba, nuestro software con base en nlas proposiciones que le damos.
