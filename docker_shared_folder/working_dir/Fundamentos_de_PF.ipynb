```scala
val fivesScala = List(5.0, 15.0, 20.0, 25.0)
```

```scala
def media(nums: List[Double]): Double = nums.sum / nums.length

def mediana(nums: List[Double]): Double = {
  val sortedNums = nums.sorted
  val middle = nums.length / 2
  if (nums.length % 2 == 0) (sortedNums(middle - 1) + sortedNums(middle)) / 2.0
  else sortedNums(middle)
}

def moda(nums: List[Double]): Double = {
  nums.groupBy(identity).mapValues(_.size).maxBy(_._2)._1
}

def calcular(nums: List[Double]): Map[String, Double] = {
  Map("Media" -> media(nums), "Mediana" -> mediana(nums), "Moda" -> moda(nums))
}
```

## Qu es un case class y en que se diferencia con la class en Scala?
un `case class` es una clase especial que se utiliza comunmente para modelar los datos inmutables. Aunque comparte muchas caractersticas con las  clases normales, **hay algunas diferencias clasve para que los case class sean utiles en escenarios especificos**
1. **Inmutabilidad por defecto**: Los `case class` son inmutables por defecto, lo que sifnidica que los objetos de este tipo no pueden ser modificados despus de que se han creado. Esto facilita el razonamietno acerca de los datos y ayuda a prevenir efectos secundarios no deseados.
2. **Generacin automtica de mtodos tiles**: Scala automticamente genera metodo tiles para las `case class`, como metodos `toString`, `equals`, y `hashCode`. Esto es til para imprimir representaciones legibles, comprar objetos y usar instancias de la clase en estructuras de datos que dependen de `hashCode`.
3. **Descomposicin de patrones**: Los case class se utilizan frecuentemente con la descomposicin de patrones en Scala. Esto permite escribir codigo conciso y expresivo para manipular y analizar la estructura de los objetos.
4. **Sintaxis concisa para la creacin de instancas**: Puedes crear instancias ``vase ckass` sin utilizar la palabra clave `new`, lo que proporciona una sintaxis ms concisa y legible.

A contuniacin, se presenta un ejemplo suimple para ilustrar la diferencia entre una clase regultasra y una `case class`;

```scala
// Clase regular
class Persona(nombre: String, edad: Int)

// Case class
case class PersonaCase(nombre: String, edad: Int)
```

Con una `case class`, automticamente obtienes mtodos como "toString", 'equals', y 'hashCode' sin tener que escribirlos explcitamente Tambin puede utilizar la descomposicin de patrones de manera ms efectiva:

```scala
val persona1 = PersonaCase("Juan", 30)

persona1 match {
  case PersonaCase("Juan", edad) => println(s"Encontr a Juan con edad $edad")
  case _ => println("No es Juan")
}
```

# Recursin
Una manera de abordar los problemas donde la funcin se llama a s misma.
Es la manera clsica de proramar en lenguajes funcionales.

## Tail recursin
Es una optimizaacin de estos algoritmos

```scala
// EL factorial con recursin

def factorial(n: Long): Long = 
    if (n == 0) {
        println("---Finalizado---")
        1
    } else {
        n * factorial(n-1)
    }
```

```scala
val startTimeRecursion = System.nanoTime()
println(factorial(100))
val endTimeRecursion = System.nanoTime()
val duration = (endTimeRecursion - startTimeRecursion) / 1e6  // Convertir a milisegundos
println(s"Tiempo de ejecucin: $duration ms")
```

## Inconvenientes solo con recursin
Hacerlo solo con recursin lo hace ms ineficiente.
En la definicin anterior, primero resta todos los numeros para al final hacer todas las multiplicacines.
Esto es ms ineficientes ya que esta multiplicacin no se va a realizar hasta que no se llegue al ltimo numero, solo hastsa que llega hasta a 1, ah arranca la multiplicacin, por lo que los datos van a quedar almacecnados en memoria.

```scala
// Tail recursion

def factorialRecursion(n: Long, resultado: Long =  1L): Long = 
    if (n == 0) {
        println("---Finalizado---")
        resultado
    } else {
        factorialRecursion(n-1, n * resultado)
    }
```

```scala
val startTimeTailRecursion = System.nanoTime()
println(factorialRecursion(100))
val endTimeTailRecursion = System.nanoTime()
val duration = (endTimeTailRecursion - startTimeTailRecursion) / 1e6  // Convertir a milisegundos
println(s"Tiempo de ejecucin: $duration ms")
```

# Razonamiento con tipos
El razonamiento inductivo es una manera de razonar donde el primer paso es encontrar lo que se conoce como caso base, es decir, uin hecho concreto que usaremos como inicio del proceso que queremos modelar.

El siguiente paso es modelar la continuacion del proceso (paso inductivo) Usando elcaso base anterior. Lo que sucede entonces es que el resultado de este proceso se convierte en el nuevo caso base.
Esto puede repetirse cuantas veces sea necesario ya que conceptualmente puede ser infinito, aunque lo normal es hallar una condicion que detiene el proceso de continuacin y entrega el resultado final.

## Ejemplo de razonamiento inductivo
Un ejemplo sencillo de una prueba por uunduccin matemtica sera probar la siguiente proposicin:
"Los numero son infinitos".
* El caso base sera el nmero ms bajo, que es el 0 (cero).
* La continuacin se usa el caso base y se le suma el numero 1.
* La prueba es que, si para cualquier caso base puedo sumarle 1, entonces los nmeros son infinitos.

El restulado de esta proposicin es que es cuerta, porque a nivel matemtico no hay nada que evite que puesa sumarle un 1 a cualquier otro nmero. Por supuesto esto no es cierto en informtica, donde estamos limitados por el espacio en memoria, esto nos pone un lmite donde ya no es posible sumar ms cuando hay n numero muy grande.

Al momento de pensar en la menera de solucionar un problema de programacin, entrenar a nuestro cerebro paar usar el rezonamieto inductivo nos puede dar soluciones ms elegantes y consistentes, lo que normalmente da como resultado software ms "correcto".

Por su puesto, no en todos los casos una solucin de esta naturaleza ser la ms eficiente computacionalmete, pero pasar de una solucin inductiva a una que no, suele ser ms facil que al revs.
